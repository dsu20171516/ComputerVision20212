### 1.4.7 Canny Edge Detection ###

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = (20, 15)
```

```python
def plt_image(img):
    channels = len(img.shape)
    if channels == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap='gray')
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.show()
```

```python
img = cv2.imread('lenna.jpg',0)

plt.subplot(231),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    
for i, x in enumerate([1, 3, 5, 7, 9]):
    blured = cv2.GaussianBlur(img, (x, x), 0)
    edges = cv2.Canny(blured, 100, 200)
    plt.subplot(int(f'23{2+i}')),plt.imshow(edges,cmap = 'gray')
    plt.title(f'kernel {x}x{x}'), plt.xticks([]), plt.yticks([])

plt.show()
```
